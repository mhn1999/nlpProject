```python
import numpy as np
import re
stopwords = [",", "یا", "تا", "اگر", "اگرچه", "پس", "که"]
def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("[A-Za-z0-9_]+"," ", temp)
    temp = re.sub("@[آ-ی]+","", temp)
    temp = re.sub("#[آ-ی]+","", temp)
    temp = re.sub(r'http.+', '', temp)
    temp = re.sub('[\[\|\\/»#\=\+:،,*\]]*','', temp)
    temp = re.sub('[()]', ' ', temp)
    temp = re.sub("\.\.+",".", temp)
    temp = re.sub("\?\?+","?", temp)
    temp = re.sub("^[0-9آ-ی]+"," ", temp)
    temp = temp.split()
    temp = [w for w in temp if not w in stopwords]
    temp = " ".join(word for word in temp)
    return temp
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("HooshvareLab/bert-base-parsbert-uncased")


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)


tokenized_datasets = train_dataset.map(tokenize_function, batched=True)
```

```python
small_train_dataset = tokenized_datasets.shuffle(seed=42).select(range(200))
small_eval_dataset = tokenized_datasets.shuffle(seed=42).select(range(200,400))
```

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch")
```

```python
trainer = Trainer(
    model=model,
    config=config,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
